# 선형 회귀 모델
: 하나의 변수를 가지는 선형회귀(직선방정식)을 위한 모델 구현

```python
import numpy as np #과학 컴퓨팅을 위한 라이브러리
import matplotlib.pyplot as plt #데이터 플로팅에 사용되는 라이브러리
```

```python
#훈련 자료(데이터) 입력(Sample data)
#array 만들기
x_train = np.array([1.0, 2.0]) # 사이즈(array - 숫자 여러개를 벡터 형태로 넣는 것(list 형태로 넣기))
y_train = np.array([300.0, 500.0]) #가격
print(f"x_train = {x_train}")#요즘 많이 쓰이는 문법(f-string) - {} 안에 변수 넣을 수 있음
print(f"y_train = {y_train}")
```

- a: scalar, non bold / **a**: vextor, bold(굵은 활자)
- **x**_train: Training Example feature values
- **y**_train: Training Example targets

```python
print(f"x_train.shape = {x_train.shape}")
m = x_train.shape[0] #m: 훈련 자료의 갯수를 나타내는 변수로 약속(Number of training examples)
#=> shape로 접근을 권장(len(x_train)보다 추천 - 파라미터를 1개만 줘서 제한적임))
#shape - 튜플로 값을 줌((2,) - numpy의 1차원 array)
print(f"Number of training example is: {m}")
```

```python
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
#데이터 넣은 거 하나씩 끄집어내보는 것
i = 0 #두 개중에서 첫 번째 element를 뽑아내는 것
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

- x_i, y_i : ith Training Example

```python
i = 1 #두 개중에서 두 번째 element를 뽑아내는 것
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
#그림 그려보기(Plotting the data)
#scatter plot - 분산 플롯 그릴 때 사용 -> x,y 좌표값만 벡터 형태로 주면 됨
#marker, color 변경 가능 - 그림 그리는 코드
plt.scatter(x_train, y_train, marker="x", c="r")#r=red / 디폴트: 파란색
plt.title("Housing prices") #title
plt.ylabel("Price(in 1000s of dollars)")#ylabel
plt.xlabel("Size (1000 sqft)")#xlabel
plt.show()
```

```python
#위의 데이터에다가 모델을 fitting - 선형 회귀 모델(직선 방정식)
#시작 파라미터:w = 100, b = 100
w = 100 #parameter: 가중치(weight) -> 100이면 그래프를 그렸을 때 맞지 않음 200으로 바꿔줘야 함
b = 100 #parameter: 편향(bias)
print(f"w: {w}")
print(f"b: {b}")
print(f"w: {w}, b:{b}")
```

```python
#함수 만들기(벡터를 다룰 수 있는 함수를 만들어야 함)
def compute_model_output(x, w, b):
    m = x.shape[0] #훈련 자료의 갯수
    f_wb = np.zeros(m) #0으로 채워진 벡터를 만드는 함수(길이가 m)/ 초깃값: 0으로 놔준 것
    for i in range(m):# 0 ~ m-1까지 하나하나 리턴
        f_wb[i] = w*x[i] + b
    
    return f_wb
```

```python
w = 200
b = 100

#함수 작동 확인하기
tmp_f_wb = compute_model_output(x_train, w, b) #임시 변수: tmp_f_wb(함수 결과값 저장)

#plot - 점들을 다 이은 line으로 plotting(scatter - 점만 찍어줌)
plt.plot(x_train, tmp_f_wb, c = "b", label = "Out predicted vlue(Our prediction)")#우리가 예측한 값: tmp_f_wb
plt.scatter(x_train, y_train, marker = "x", c = "r", label = "Actual(actual)")#원래 값

plt.title("Housing Prices")
plt.ylabel("price(in 100s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.legend()#라벨에 맞춰서 알아서 symbol을 그려줌
plt.show()
```

```python
#1200제곱피트의 주택 가격 예측
#이런 작업을 자동적으로 해야됨 => 선형회귀가 하는 것
w = 200
b = 100

x_i = 1.2
cost = w * x_i + b
print(f"${cost:.0f} thousand dollars") # .0f - 소수점 아래는 자르기(정수로 써달라는 표현)
```
